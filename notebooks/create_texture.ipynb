```python
import numpy as np

from matplotlib.image import imread, imsave
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img_heightmap_path = "../models/terrain/terrain-heightmap.png"
img_texture_path = "../models/terrain/terrain-texture.png"
```

```python
img_heightmap = imread(img_heightmap_path)
plt.imshow(img_heightmap);
img_heightmap.shape
```

```python
img_texture = np.zeros(img_heightmap.shape)

s1 = 1*257//5
s2 = 2*257//5
s3 = 3*257//5
s4 = 4*257//5

# Color high-mobility terrain red
img_texture[np.r_[:s1, s2:s3, s4:257], :, 0] = 1

# Color medium-mobility terrain green
img_texture[s1:s2, :, 1] = 1

# Color low-mobility terrain blue
img_texture[s3:s4, :, 2] = 1

plt.imshow(img_texture);
```

```python
plt.figure(figsize=(10, 10))

plt.imshow(img_heightmap)
plt.imshow(img_texture, alpha=0.1);
```

```python
plt.imsave(img_texture_path, img_texture)
```
